````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import imageio
````

# Uploading a black image

````python
img = imageio.imread('cat-bw.png')
````

````python
# height, width, number of channels
print(img.shape)
````

````python
img_bw = img[:, :, 0]
print(img_bw.shape)
````

# Image visualisation

````python
# Working Environment
plt.figure(figsize=(8, 8))
plt.imshow(img_bw, cmap='gray');
````

````python
# numpy displays it as a tensor
img_bw
````

# Image cropping

````python
img_crop = img_bw[100:300, 200:500]
print(img_crop.shape)
````

````python
plt.figure(figsize=(8, 8))
plt.imshow(img_crop, cmap='gray')
````

````python
plt.imshow(img_bw[100:105, 200:205], cmap='gray')
````

# Changing the image

````python
img_crop[80:130, 70:120] = 20
````

````python
plt.figure(figsize=(8, 8))
plt.imshow(img_crop, cmap='gray')
````

````python
# we put in the main picture as we did not make a copy
plt.figure(figsize=(8, 8))
plt.imshow(img_bw, cmap='gray')
````

# Copying the image

````python
img_crop = np.copy(img_bw[300:500, 200:500])
print(img_crop.shape)
````

````python
plt.figure(figsize=(8, 8))
plt.imshow(img_crop, cmap='gray')
````

````python
img_crop[80:130, 70:120] = 20
````

````python
plt.figure(figsize=(8, 8))
plt.imshow(img_crop, cmap='gray')
````

````python
# It's fine now, the original hasn't changed
plt.figure(figsize=(8, 8))
plt.imshow(img_bw, cmap='gray')
````

# Uploading a colour image

````python
img = imageio.imread('cat-color.png')
print(img.shape)
````

```shell
The data is stacked in triples, according to the number of RGB channels.
The first index is the slowest to change, while the last index is the fastest.
```

````python
print(img[:5, :5, :])
````

# Visualising a colour image

````python
plt.figure(figsize=(8, 8))
plt.imshow(img)
````

# Visualising the image channels

````python
plt.figure(figsize=(8, 8))
plt.imshow(img[:, :, 0], cmap='gray') # Red
````

````python
plt.figure(figsize=(8, 8))
plt.imshow(img[:, :, 1], cmap='gray') # Green
````

````python
plt.figure(figsize=(8, 8))
plt.imshow(img[:, :, 2], cmap='gray') # Blue
````

# Pixel colour

````python
print(img[100, 200, :]) # RGB
````

# Conversion to float32 [0,1]

````python
img = img.astype(np.float32) / 255.
````

````python
img[:5, :5, :]
````

````python
plt.figure(figsize=(8, 8))
plt.imshow(img)
````
