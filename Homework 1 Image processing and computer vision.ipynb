For my homework, I chose to implement a sepia-colored image that mimics an old black and white photo.

To do this, each pixel of the original image must be modified according to the following rule:
```shell
outputRed = (inputRed * .393) + (inputGreen *.769) + (inputBlue * .189)
outputGreen = (inputRed * .349) + (inputGreen *.686) + (inputBlue * .168)
outputBlue = (inputRed * .272) + (inputGreen *.534) + (inputBlue * .131)
```

````python
import matplotlib.pyplot as plt
import numpy as np

import imageio
````

# Uploading a colour image

````python
img = imageio.imread('cat-color.png')
img = img.astype(np.float32) / 255.
print(img.shape)
````

````python
plt.figure(figsize=(8, 8))
plt.imshow(img);
````

````python
# create a picture (initialise the buffer to be filled)
# dimension 3: height, width, number of channels, with data type float32
out_img = np.zeros(
            (img.shape[0],
             img.shape[1],
             img.shape[2]), 
             dtype=np.float32)

# go through the height and width
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        
        # store the initial values of the red, green and blue channels of each pixel
        input_red = img[i, j][0]
        input_green = img[i, j][1]
        input_blue = img[i, j][2]
        
        # Calculate the RGB value for the current pixel using the formula 
        output_red = (input_red * .393) + (input_green *.769) + (input_blue * .189)
        output_green = (input_red * .349) + (input_green *.686) + (input_blue * .168)
        output_blue = (input_red * .272) + (input_green *.534) + (input_blue * .131)

        # write the new RGB pixel values with a check that they are greater than 1
        out_img[i, j] = [output_red if output_red < 1 else 1, 
                         output_green if output_green < 1 else 1, 
                         output_blue if output_blue < 1 else 1
                        ]
        
plt.figure(figsize=(8, 8))
plt.imshow(out_img);
````
